```python
import pandas as pd
```

```python
# pd.read_csv("BaseJsonCidades.txt", encoding="ISO-8859-1")
lista=[]
for line in open('BaseJsonCidades.txt', encoding = "ISO-8859-1"):
    lista.append(line)

lista = lista[5828:] #onde comeÃ§a a ultima rodagem valida que eu fiz para o segundo turno
```

```python
import json
# with open('BaseJsonCidades.txt') as f:
#     json_data = json.load(f)

df7 = pd.DataFrame() #dataframe em branco onde vai ser armazenado tudo

for i in range(len(lista)):
    if i%2>0:
        x = lista[i].replace("'",'"')
        json_data = json.loads(x)
        df = pd.DataFrame(json_data).drop("abr",axis=1)
        df2 = pd.DataFrame(json_data['abr'])
        df3 = df.join(df2)
        
        #iterar sobre as zonas da cidade
        for n in range(len(json_data['abr'])):

            df4 = pd.DataFrame(df3.drop('cand',axis=1).iloc[n]).transpose().reset_index()
            df5 = pd.DataFrame(df3['cand'][n])
            # pd.merge(df5,df4, how="outer")
            df6 = pd.merge(df5, df4, left_index=True, right_index=True, how="outer").fillna(method="ffill")
            df6['cdabr_cod_cidade'] = json_data['abr'][0]['cdabr'] #sempre zero
            df7 = pd.concat([df6,df7])

        
```
